```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f,LearningRate, InitialPoint): #estimates derivative of the function f at base point
    def derivative(f, BasePoint): #Intakes a function and a point
        return (f(BasePoint)+10**(-5))-(f(BasePoint)/(10**(-5))) #Calculates the derivative
    x_coords=[InitialPoint] #stores x values
    y_coords=[f(InitialPoint)] #stores y values
    
    Point=(f(InitialPoint-LearningRate*derivative(f,InitialPoint))) #Initializes point as the function of the input value - learning rate multiplied by the derivative of the function
    for i in range(1000): #if no minimum found, either increase step size or tell user no dice
        x_coords.append(Point) #Adds x-value to list
        y_coords.append(f(Point)) #Adds y-value to list
        Point=Point-LearningRate*derivative(f,Point) #subsequent points change here from x1, x2... xn
        if derivative(f,Point) < 0.001: #Error bound
            PlotRange=np.linspace(min(x_coords)-0.5,max(x_coords)+0.5,10000) #Plots the coordinates
            FunctionRange=[f(i) for i in PlotRange] #Range of function
            plt.plot(PlotRange,FunctionRange) #plots the function f(x)
            plt.plot(x_coords,y_coords) #This will plot the sequence of points x_n, f(x_n)
            return round(x_coords[-1],3), round(y_coords[-1],3) #returns last x_n and y_n; rounds to 3 decimal places
                                                            
```

```python
f = lambda x: x**2 
```

```python
gradient_descent(f,0.02,1)
```

```python
f = lambda x: -3*(x**3)+10*(x**2)-(5*x)-3
```

```python
gradient_descent(f,0.2,1)
```

```python
def f(x):
    if x >0:
        return x**x
    elif x == 0:
        return 1
    else:
        return abs(x)**abs(x)
```

```python
gradient_descent(f,0.1,1)
```

```python
gradient_descent(f,0.1,1)
```
