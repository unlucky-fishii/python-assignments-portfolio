```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = sp.symbols('x y')
# Define a function
f = x**2 + sp.sin(y)
# Differentiate f with respect to x
df_dx = sp.diff(f, x)
# Differentiate f with respect to y
df_dy = sp.diff(f, y)
```

```python
#Excersize A
g = sp.exp(x)*sp.sin(y)+y**3
#Differentiate f with respect to x
dg_dx = sp.diff(g,x)
#Differentiate f with respect to y
dg_dy = sp.diff(g,y)
```

```python
#Excersize B; #(1,-1)
h = x**2*y + x*y**2
#Gradient vector (Du(f)) = df/dx, df/dy)
dh_dx = sp.diff (h,x)
dh_dy = sp.diff(h,y)

Du = (dh_dx,dh_dy)
print(Du)

print(Du[0].subs([(x,1),(y,-1)]),Du[1].subs([(x,1),(y,-1)]))
```

```python
#Excersize C
k = sp.log(x**2+y**2)

dk_dx = sp.diff(k,x)
print(dk_dx)

dk_dy = sp.diff(k,y)
print(dk_dy)

dk_dxx = sp.diff(dk_dx,x)
print(dk_dxx)

dk_dyy = sp.diff(dk_dy,y)
print(dk_dyy)

dk_dxy = sp.diff(dk_dx,y)
print(dk_dxy)

dk_dyx = sp.diff(dk_dy,x)
print(dk_dyx)

#dk_dxy = dk_dyx in general. 
```

```python
#Excersize D
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)
plt.contourf(X, Y, Z, levels=10, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#Excersize E
x, y = sp.symbols('x y')
j = x**2*y + x*y**2
j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)
plt.contourf(X, Y, Z, levels=10, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^2y + xy^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
